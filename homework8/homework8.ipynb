### maximum, minimum, mean age by gender before wreck of the Titanic

```python
data_train_age =  data_train.groupby('Sex')['Age'].agg([('min_age','min'), ('max_age','max'), ('mean_age','mean'), ('number before the wreck','count')]).reset_index()
data_train_age
```

```python
data_passanger_survived = data_train[data_train.Survived == 1]
data_passanger_survived
```

### Number of survivors

```python
passanger_survived = data_passanger_survived['Age'].count()
passanger_survived
```

```python
count_age_isnull = data_train[data_train.Age.isnull()]['PassengerId'].count()
count_age_isnull
```

### Maximum, minimum, mean age of survivors by gender after wreck of the Titanic

```python
data_passanger_survived_age = data_passanger_survived.groupby('Sex')['Age'].agg([('min_age','min'), ('max_age','max'), ('mean_age','mean'), ('survived','count')]).reset_index()
data_passanger_survived_age
```

```python
data_for_graph = data_passanger_survived_age.merge(data_train_age, on='Sex')
```

```python
plt.rcParams["figure.figsize"] = (5,5)
graph = data_for_graph.plot(x='Sex', y = ['number before the wreck', 'survived'], kind = 'bar')
graph.locator_params(axis='y', nbins= 10)
```
